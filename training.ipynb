# Model Training

```python
#Baseline logistic regression model
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import (
    roc_auc_score, classification_report, confusion_matrix,
    roc_curve, auc
)
import matplotlib.pyplot as plt

df = pd.read_csv("Train_Dataset.csv")

RANDOM_STATE = 42
TARGET_COL = "Default"
#ID is not an informative feature
DROP_COLS = ["ID"]

# 1) Preprocess
df_lr = df.copy()

# Convert numeric-looking strings to actual numbers
for col in df_lr.columns:
    if df_lr[col].dtype == "object":
        conv = pd.to_numeric(df_lr[col], errors="coerce")
        if conv.notna().mean() > 0.10:
            df_lr[col] = conv

# Force these to be treated as categorical
df_lr["Cleint_City_Rating"] = df_lr["Cleint_City_Rating"].astype("object")
df_lr["Application_Process_Day"] = df_lr["Application_Process_Day"].astype("object")

# Seperate features from target in dataframe

X = df_lr.drop(columns=[c for c in [TARGET_COL] + DROP_COLS if c in df_lr.columns])
y = df_lr[TARGET_COL].astype(int)

# split X dataframe into numerical and categorical columns
numeric_cols = X.select_dtypes(include=["int64", "float64"]).columns
categorical_cols = X.select_dtypes(include=["object"]).columns

# put any NaN entry as the median of the column
X_num = X[numeric_cols].apply(pd.to_numeric, errors="coerce")
X_num = X_num.fillna(X_num.median(numeric_only=True))

# One-hot encode categoricals

ohe = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
X_cat = ohe.fit_transform(X[categorical_cols].astype(str).fillna("Missing"))

# Recombine
X_final = np.hstack([X_num.to_numpy(), X_cat])

# Scale all features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_final)
```

# Takeaway 

Data frame cleaning and one hot encoding ready to be used for the logistic regression baseline model

```python
# 2) Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)

# 3) Fit Logistic Regression
log_reg = LogisticRegression(max_iter=5000, random_state=RANDOM_STATE)  # try class_weight="balanced" later
log_reg.fit(X_train, y_train)

# 4) Evaluate

#training set (threshold = 0.5)
proba_train = log_reg.predict_proba(X_train)[:, 1]
pred_train  = (proba_train >= 0.50).astype(int)
# test set
proba_test = log_reg.predict_proba(X_test)[:, 1]
pred_test  = (proba_test >= 0.50).astype(int)

# ROC Curve analysis: computes for many thresholds fpr and tpr and numerically integrates the curve formed by these
#points to determine AUC score
fpr_train, tpr_train, _ = roc_curve(y_train, proba_train)
roc_auc_train = auc(fpr_train, tpr_train)
fpr_test, tpr_test, thresholds = roc_curve(y_test, proba_test)
roc_auc_test = auc(fpr_test, tpr_test)

plt.figure(figsize=(6, 5))
plt.plot(fpr_train, tpr_train, label=f"Train ROC (AUC={roc_auc_train:.3f})", linewidth=2)
plt.plot(fpr_test, tpr_test, label=f"Test ROC (AUC={roc_auc_test:.3f})", linewidth=2)
plt.plot([0, 1], [0, 1], linestyle="--", linewidth=1, label="Chance")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate (Recall)")
plt.title("ROC Curve — Logistic Regression")
plt.legend(loc="lower right")
plt.tight_layout()
plt.show()

# Summary metrics (TEST)
roc = roc_auc_score(y_test, proba_test)
print(f"Test ROC-AUC: {roc:.4f}\n")
print("Classification Report (threshold=0.50):")
print(classification_report(y_test, pred_test))
print("Confusion Matrix (threshold=0.50):")
print(confusion_matrix(y_test, pred_test))

#tpr/fpr at specific thresholds
def tpr_fpr_at_threshold(y_true, proba, thr):
    pred = (proba >= thr).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_true, pred).ravel()
    tpr = tp / (tp + fn) if (tp + fn) else 0.0
    fpr = fp / (fp + tn) if (fp + tn) else 0.0
    return tpr, fpr

print('------------------------')
for thr in np.arange(0,1,0.1):
    tpr, fpr = tpr_fpr_at_threshold(y_test, proba_test, thr)
    print(f"Threshold={thr:.2f} -> TPR={tpr:.3f}, FPR={fpr:.3f}")
print('------------------------')
```

# Takeaway 

Run logistic regression. Show performance by comparing the ROC AUC curves for the training and testing sets

```python
# understand relationship between ROC and regularization parameter
for C in [1e-8, 1e-7, 1e-6, 1e-5, 1e-4 , 0.01, 0.1, 1, 10, 100000000000000]:
    lr_C = LogisticRegression(max_iter=5000, random_state=RANDOM_STATE, C=C)
    lr_C.fit(X_train, y_train)

    probaC_train = lr_C.predict_proba(X_train)[:, 1]
    probaC_test = lr_C.predict_proba(X_test)[:, 1]

    train_auc = roc_auc_score(y_train, probaC_train)
    test_auc = roc_auc_score(y_test, probaC_test)
    diff_auc = train_auc-test_auc
    print(f"C={C:<4}  Train ROC-AUC={train_auc:.4f} Test ROC-AUC={test_auc:.4f} Diff={diff_auc:.4f}")

# -- Hyperparameter Tuning --
from sklearn.model_selection import GridSearchCV, cross_val_score
#param_grid is a matrix of hyperparamters, we would like to know what is the most optimal of the 6*1*2 = 12 configurations
param_grid = {
    "C": [1e-4, 0.01, 0.1, 1, 10, 100],
    "penalty": ["l2"],
    #class weight is a hyperparamter. "balanced" will weigh minority labels as more important
    "class_weight": [None, "balanced"]
}

grid_search = GridSearchCV(
    LogisticRegression(max_iter=5000, random_state=RANDOM_STATE, solver="lbfgs"),
    param_grid=param_grid,
    scoring="roc_auc",
    #split into 80% training and 20% CV, splits into 5 chunks and each chunk is a cross validation set once.
    cv=5,
    n_jobs=-1,
    verbose=2
)
#runs the search fit

#for each choices of hyperparamters, trains the model
grid_search.fit(X_train, y_train)

#output what it found to be the best parameters and its average auc score across the 5 CV sets.
print("Best parameters:", grid_search.best_params_)
print("Best mean CV ROC-AUC:", grid_search.best_score_)

#with this best chosen hyperparameters, we proceed to report the AUC of the logistic regression model on the testing set
best_lr = grid_search.best_estimator_
proba_best = best_lr.predict_proba(X_test)[:, 1]
auc_best = roc_auc_score(y_test, proba_best)
print(f"Tuned Test ROC-AUC: {auc_best:.4f}")
```

# Takeaway
* Training logistic regression with different regularization parameters and evaluating performance using ROC-AUC score. 

* Hyperparameter tuning using a grid with 12 combinations of hyperparametes and finding the best combination to output an unbiased test AUC score

```python
# --- Standard Decision Tree (baseline) ---
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
# 1) Train a decision tree
# basic decision tree first

X =X_num.copy()

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)

dt = DecisionTreeClassifier(random_state=RANDOM_STATE,max_depth=8)
dt.fit(X_train, y_train)
# basic decision tree first

# 2) Test-set predictions
proba_test_dt = dt.predict_proba(X_test)[:, 1]   # probabilities for ROC/AUC
pred_test_dt  = dt.predict(X_test)
#From proportion of defaults in a lead during training, assigns a sample that lands in that leaf to that computed probability.
#Based on a threshold, (default=0.5), pred_test_dt is a array where arr[i] is whether the ith leaf classifies 0 or 1.

# 3) Metrics on TEST
auc_dt = roc_auc_score(y_test, proba_test_dt)
print(f"Decision Tree — Test ROC-AUC: {auc_dt:.4f}\n")

print("Classification Report (thr=0.50):")
print(classification_report(y_test, pred_test_dt))

print("Confusion Matrix (thr=0.50):")
print(confusion_matrix(y_test, pred_test_dt))

# 4) ROC curve (TEST)
fpr_dt, tpr_dt, _ = roc_curve(y_test, proba_test_dt)

plt.figure(figsize=(6,5))
plt.plot(fpr_dt, tpr_dt, label=f"Decision Tree (AUC={auc_dt:.3f})", linewidth=2, marker='*')
plt.plot([0,1],[0,1],"--",color="gray", label="Chance")
plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate (Recall)")
plt.title("ROC — Decision Tree (Test Set)")
plt.legend(loc="lower right"); plt.tight_layout(); plt.show()
```
